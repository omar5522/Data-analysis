```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.jumia.com.eg/ar/flash-sales/'
```

```python
# Set of visited URLs to avoid duplicate scraping
visited_urls = set()

# Set of seen document content to avoid repeating output
seen_doc_content = set()
```

```python
response = requests.get(url)
html = response.content
soup = BeautifulSoup(html,'html.parser')
print(soup.prettify())
```

```python
main_content = soup.find('div', {'class': '-pvs col12'})
```

```python
if main_content is None:
    print("None")
        # No main content section found, skip scraping this page
```

```python

doc_content_list = [element.text.strip() for element in main_content.find_all(['h3'])]
doc_content_list
```

```python
price = soup.find_all('div',{'class':'prc'})
price_list = [element.text.strip() for element in price]
price_list
```

```python
data = {'اسم المنتج': doc_content_list, 'سعر المنتج': price_list}

# Create the DataFrame
df = pd.DataFrame(data)

# Save the DataFrame to an Excel file
df.to_excel('New.xlsx', index=False)
```
